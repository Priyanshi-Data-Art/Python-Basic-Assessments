```python
## explain the concept of type casting in Python with examples.
In Python, Type casting is used to convert one data type to another. There are two types of typecasting i.e explicit type casting and
implicit typecasting.

Implicit Typecasting:- Python automatically change one data type to another according to their need that is called implicit type casting.

For Example:-
a = 5  # interger
b = 5.5 # float

c = a+b 
print(c) #it'll give 10.5 i.e float

Explicit Typecasting:- When we mannualy change data type with the help of python built-in-function that is known as explicit typecasting.

To convert any data-type to Integer, we use int() function.

For Example:-
a = '100'

print(int(a)) #it will convert string to integer

To convert any data-type to string, we use str() function.

for Example:-

b = 100

print(str(b)) #it will convert integer to String

To convert any data-type to float, we use float() function.

for Example:-

b = 100

print(float(b)) #it will convert integer to float
```

```python
#How do conditional statements work in Python? Illustrate with examples.

In Python, when some piece of code execute based on whether condition is True or False this is known as conditional statement.
if result is true then code will execute otherwise it will not execute.

Types of Conditional Statement:-
'if' -> when 'if' condition is True then block of code will execute

For example:-

age = 19

if age >= 18:
    print("You are eligible to go college")
    
'if-else' -> when if condition is True control will go to if block of statement, otherwise control will pass to else block of code.

color = blue

if color == blue:
    print("It's my fav color")
else:
    print("It's not my fav color")
    
'if-elif-else' -> This allows to multiple condition. The first condition that evaluates to True will have its block executed, and 
the rest will be skipped.

marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 80:
    print("Grade: B")
elif marks >= 70:
    print("Grade: C")
else:
    print("Grade: D")

'Nested if' -> In this, If statement place inside another if statement.

age = 20
citizenship = "Yes"

if age >= 18:
    if citizenship == "Yes":
        print("You are eligible to vote.")
    else:
        print("You must be a citizen to vote.")
else:
    print("You are not old enough to vote.")
```

```python
## Describe the different types of loops in Python and their use cases with examples.

In Python, loops are used to execute a block of code repeatedly as long as a certain condition is met. There are three main types of loops in Python:-

'for' -> 'for' loops are used for iterating over a sequence of elements.

code:-
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

Use-Case:- It is used to perform an action a specific number of times.

'while' -> “while” loops are used for executing code until a specified condition becomes False.

code:-
count = 0
while count < 5:
    print("Count:", count)
    count += 1

Use-Case:- Repeating actions until a certain condition changes.
```
